```python
import pandas as pd
import faker
```

```python
# Genera un dataframe ficticio con faker que contenga logs de un servidor
# con info semi estructurada,
# una sola columna, con registrosç delimitados
def generate_fake_logs(n):
    fake = faker.Faker()
    logs = []
    for _ in range(n):
        ip = fake.ipv4()
        tstamp = fake.date_time_this_year()
        method = fake.random_element(["GET", "POST", "PUT", "DELETE"])
        url = fake.uri()
        request = f"{method} {url}"
        status = fake.random_int(100, 599)
        user_agent = fake.user_agent()
        bytes = fake.random_int(100, 10000)
        logs.append(f'{ip} - {tstamp} - "{request}" - {status} - {bytes} - "{user_agent}"')

    return pd.DataFrame(logs, columns=["log"])

df = generate_fake_logs(1000)
```

```python
pd.set_option('display.max_colwidth', None)
```

```python
df.head()
```

```python
df_structured = df["log"].str.split(" - ", expand=True, n=5)
# renombrar columnas
df_structured = df_structured.rename(
    columns={
        0: "ip",
        1: "tstamp",
        2: "request",
        3: "status",
        4: "bytes",
        5: "user_agent"
    }
)
df_structured.head()
```

```python
rename_mapping = {
    "ip": "ip_cliente",
    "tstamp": "fecha_hora",
    "request": "peticion",
    "status": "codigo_estado",
    "bytes": "bytes_transmitidos",
    "user_agent": "navegador_cliente"
}
df_structured = df_structured.rename(columns=rename_mapping)
```

```python
df_structured.head()
```

```python
df_structured[["metodo", "url"]] = (df_structured["peticion"]
                                    .str.strip('"')
                                    .str.split(" ", expand=True, n=1)
)
```
